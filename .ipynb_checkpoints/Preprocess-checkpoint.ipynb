```python
SAVEE = r'E:\datasets\savee'
RAVDESS = r'E:\datasets\ravdess'
CREMA = r'E:\datasets\crema'
TESS = r'E:\datasets\tess'
```

# SAVEE

```python
savee_directory_list = os.listdir(SAVEE)
file_emotion = []
file_path = []

for file in savee_directory_list:
    file_path.append(os.path.join(SAVEE, file))
    part = file.split('_')[1]
    ele = part[:-6]
    if ele=='a':
        file_emotion.append('angry')
    elif ele=='d':
        file_emotion.append('disgust')
    elif ele=='f':
        file_emotion.append('fear')
    elif ele=='h':
        file_emotion.append('happy')
    elif ele=='n':
        file_emotion.append('neutral')
    elif ele=='sa':
        file_emotion.append('sad')
    else:
        file_emotion.append('surprise')
        
# dataframe for emotion of files
emotion_df = pd.DataFrame(file_emotion, columns=['Emotions'])

# dataframe for path of files.
path_df = pd.DataFrame(file_path, columns=['Path'])
Savee_df = pd.concat([emotion_df, path_df], axis=1)
Savee_df["Gender"] = "male"
Savee_df["Source"] = 'savee'
Savee_df.head()
# Savee_df.to_csv("savee.csv", index=False)
```

# RAVDESS

```python
ravdess_directory_list = os.listdir(RAVDESS)

file_emotion = []
file_gender = []
file_path = []
for dir in ravdess_directory_list:
    # as their are 20 different actors in our previous directory we need to extract files for each actor.
    actor = os.listdir(RAVDESS+'\\' + dir)
    for file in actor:
        part = file.split('.')[0]
        part = part.split('-')
        # third part in each file represents the emotion associated to that file.
        file_emotion.append(int(part[2]))
        file_gender.append(int(part[-1]))
        file_path.append(os.path.join(RAVDESS,dir,file))
        
# dataframe for emotion of files
emotion_df = pd.DataFrame(file_emotion, columns=['Emotions'])
gender_df = pd.DataFrame(file_gender, columns = ['actor'])
gender_df['Gender'] = np.where(gender_df['actor'] % 2 != 0, 'male', 'female')

# dataframe for path of files.
path_df = pd.DataFrame(file_path, columns=['Path'])
Ravdess_df = pd.concat([emotion_df, path_df, gender_df], axis=1)

# changing integers to actual emotions.
Ravdess_df.Emotions.replace({1:'neutral', 2:'calm', 3:'happy', 4:'sad', 5:'angry', 6:'fear', 7:'disgust', 8:'surprise'}, inplace=True)
Ravdess_df['Source'] = 'ravdess'
Ravdess_df.head()
# Ravdess_df.to_csv("ravdess.csv", index=False)
```

# CREMA-D

```python
crema_directory_list = os.listdir(CREMA)

file_emotion = []
file_gender = []
file_path = []

for file in crema_directory_list:
    # storing file paths
    file_path.append(os.path.join(CREMA,file))
    # storing file emotions
    part=file.split('_')
    if part[2] == 'SAD':
        file_emotion.append('sad')
    elif part[2] == 'ANG':
        file_emotion.append('angry')
    elif part[2] == 'DIS':
        file_emotion.append('disgust')
    elif part[2] == 'FEA':
        file_emotion.append('fear')
    elif part[2] == 'HAP':
        file_emotion.append('happy')
    elif part[2] == 'NEU':
        file_emotion.append('neutral')
    else:
        file_emotion.append('Unknown')
    file_gender.append(int(part[0]))
        
# dataframe for emotion of files
emotion_df = pd.DataFrame(file_emotion, columns=['Emotions'])
gender_df = pd.DataFrame(file_gender, columns = ['actor'])

gender = []
female_ids = [1002,1003,1004,1006,1007,1008,1009,1010,1012,1013,1018,1020,1021,
              1024,1025,1028,1029,1030,1037,1043,1046,1047,1049,1052,1053,1054,
              1055,1056,1058,1060,1061,1063,1072,1073,1074,1075,1076,1078,1079,
              1082,1084,1089,1091]

gender_df['Gender'] = 'male'
actor_s = gender_df['actor']
rows = []
for i in range(gender_df.shape[0]):
    row = gender_df.iloc[i]
    if row['actor'] in female_ids:
        rows.append(i)
gender_df.Gender.iloc[rows] = 'female'
# gender_df['Gender'].value_counts()

# dataframe for path of files.
path_df = pd.DataFrame(file_path, columns=['Path'])
Crema_df = pd.concat([emotion_df, path_df, gender_df], axis=1)
Crema_df['Source'] = 'crema'
Crema_df.head()
# Crema_df.to_csv("crema.csv", index=False)
```

# TESS

```python
tess_directory_list = os.listdir(TESS)

file_emotion = []
file_path = []

for dir in tess_directory_list:
    directories = os.listdir(TESS+'\\' + dir)
    for file in directories:
        part = file.split('.')[0]
        part = part.split('_')[2]
        if part=='ps':
            file_emotion.append('surprise')
        else:
            file_emotion.append(part)
        file_path.append(os.path.join(TESS,dir,file))
        
# dataframe for emotion of files
emotion_df = pd.DataFrame(file_emotion, columns=['Emotions'])

# dataframe for path of files.
path_df = pd.DataFrame(file_path, columns=['Path'])
Tess_df = pd.concat([emotion_df, path_df], axis=1)
Tess_df["Gender"] = "female"
Tess_df['Source'] = 'tess'
Tess_df.head()
# Tess_df.to_csv("tess.csv", index=False)
```

```python
import pandas as pd
import numpy as np
import sklearn
from scipy import stats
import matplotlib.pyplot as plt
import os
import seaborn as sns
import librosa
import librosa.display
import random

data_path = pd.read_csv('base_data.csv')

plt.title('Count of Emotions', size=16)
sns.countplot(data_path.Emotions)
plt.ylabel('Count', size=12)
plt.xlabel('Emotions', size=12)
sns.despine(top=True, right=True, left=False, bottom=False)
plt.show()
```

```python
def create_waveplot(data, sr, e, path):
    plt.figure(figsize=(10, 3))
    plt.title('Waveplot: {} [{}]'.format(path,e), size=15)
    librosa.display.waveplot(data, sr=sr)
    plt.show()

def create_spectrogram(data, sr, e, path):
    # stft function converts the data into short term fourier transform
    X = librosa.stft(data)
    Xdb = librosa.amplitude_to_db(abs(X))
    plt.figure(figsize=(12, 3))
    plt.title('Spectrogram: {} [{}]'.format(path, e), size=15)
    librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='hz')   
    #librosa.display.specshow(Xdb, sr=sr, x_axis='time', y_axis='log')
    plt.colorbar()
    plt.show()
    
def plot(data, sr, e, path):
    create_waveplot(data, sr, e, path)
    create_spectrogram(data, sr, e, path)
```
